**One-Hot Representation**<br>
Based on a Hash function and a vocabulary size decided by us, one_hot function assigns an index number to each word. It implies that we will have a binary one_hot vector for each word where each entry is 0 except the mentioned index position where value would be 1.<br>
Such one hot reperesention doesn't capture the sematic relationship between words. For that, we use Word2Vec, Embedding layer concepts etc.

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
corp = ['Hi nice to meet you', 'You seem quite interesting', 'What do you do', 'What brought you here', 'Would you care to have a cup of coffee']
```

```python
vocab_size = 100
new_corp = [one_hot(sent, vocab_size) for sent in corp]
```

```python
new_corp
```

```python
m_len = max([len(x) for x in new_corp])
m_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_corp = pad_sequences(new_corp, maxlen = m_len, padding = 'pre')
padded_corp
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
```

```python
feature_vec = 20
model = Sequential()
model.add(Embedding(vocab_size, feature_vec, input_length = m_len))
model.compile(loss ='mse',optimizer = 'adam')
```

```python
model.summary()
```

```python
model.predict(padded_corp)
```
