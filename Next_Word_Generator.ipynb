```python
#delimiter = '\n'
with open('spacetime.txt', 'r') as file:
  spt = file.read()
```

```python
spt
```

```python
spt_data = []
#for i in range(len(spt)):
ssp = re.sub('[^.a-zA-Z]', ' ', spt)
ssp = re.sub('\s+', ' ', ssp)   #removing unnecessary spaces
  #ssp = re.sub('\n\s*\n', '.', ssp)
spt_data.append(str(ssp))
spt_data
```

```python
sp_new = spt_data[0].split(' ')
sp_new
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts([sp_new])
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
for sentence in spt_data[0].split('.'):
  print(tokenizer.texts_to_sequences([sentence])[0])
```

```python
input_seq = []
for sentence in spt_data[0].split('.'):
  tokenized_sent = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1, len(tokenized_sent)):
    input_seq.append(tokenized_sent[:i+1])
```

```python
input_seq
```

```python
max_len = max([len(x) for x in input_seq])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_seq = pad_sequences(input_seq, maxlen = max_len, padding = 'pre')
```

```python
padded_input_seq
```

```python
X = padded_input_seq[:,:-1]
```

```python
y = padded_input_seq[:,-1]
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes = 479)   #479 because len(tokenizer.word_index) + 1; +1 because one-hot-encodin starts from 0.
```

```python
y
```

```python
y.shape
```

Now **Architecture**<br>
Embedding layer -> LSTM -> Dense Layer(Fully Connected layer)<br>
for one sentence, input to **Embedding layer** is (1,63) because 63 words in each sentence and output is (1,63,100) because each word is cnverted into 100 dimensional vector here.<br>
for **LSTM layer**, there will be 63 time steps (because 63 words in each sentence) and each time step will feed a 100 dimensional input word. LSTM will provide a 150 dimesional output(a hyperparameter).<br>
Dense Layer has 479 nodes becasue vocab size is 478 and final output shape is also required to be of 479 dimensions. Hence Dense layer will perform $σ$(WX + b) on all 479 nodes to give a probability vector of size 479. The highest probability index will give us desired word as prediction.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(479, 100, input_length = 62))
model.add(LSTM(150))
model.add(Dense(479, activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y, epochs = 80)
```

```python
text = ' same events'
#for next word prediction, first tokenize then do padding then do prediction
text_token = tokenizer.texts_to_sequences([text])[0]
```

```python
text_token
```

```python
padded_text_token = pad_sequences([text_token], maxlen = 62, padding = 'pre')   #here pay attention that maxlen took 62 but earlier maxlen took max_len i.e. 63 because earlier target y column is not truncated there.
print(padded_text_token)
```

```python
pos = np.argmax(model.predict(padded_text_token))
pos
```

```python
for word, index in tokenizer.word_index.items():
  if index == pos:
    print(word)
    break
```

```python
for i in range(6):
  token_text = tokenizer.texts_to_sequences([text])[0]
  padded_token_text = pad_sequences([token_text], maxlen = 62, padding = 'pre')
  pos = np.argmax(model.predict(padded_token_text))

  for word, index in tokenizer.word_index.items():
    if pos == index:
      text = text + ' ' + word
      print(text)
      break
```
